```python
from optiflow import define_matrices, compute_displacements_from_kymograph,plot_vector_field
import numpy as np 
import matplotlib.pyplot as plt
```

## 1 : Load Kymograph

```python
K = np.load("Kymograph_example.npy")
plt.figure()
plt.imshow(K,plt.cm.magma,extent = (0,1,0,1))
plt.xticks([], [])
plt.yticks([], [])
plt.ylabel("Spatial dimension")
plt.xlabel("Time dimension")
plt.colorbar()
plt.title("Kymograph")

plt.figure()
plt.plot(np.mean(K,axis=0))
plt.ylim(0,np.amax(K))
plt.xlabel("Time dimension")
plt.ylabel("Intensity")
plt.title("Evolution of the mean concentration across time")
```

## 2 : Define Optimal Transport Matrices

```python
C,D = define_matrices(npoints = len(K),plot=True)
```

## 3 : Compute displacements with Optimal Transport

```python
V = compute_displacements_from_kymograph(K,C,D)
```

```python
vmin = min(np.abs(V.min()),V.max())
plt.imshow(V.clip(-vmin,vmin),plt.cm.bwr,extent = (0,1,0,1))
plt.xticks([], [])
plt.yticks([], [])
plt.ylabel("Spatial dimension")
plt.xlabel("Time dimension")
plt.colorbar()
plt.title("Velocity field")
```

## 4 : Display Vector field with polyscope

```python
plot_vector_field(V,t=7,pool_size = 25)
```
